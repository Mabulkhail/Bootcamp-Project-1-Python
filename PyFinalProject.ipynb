# Casher System

```python
customer_info = {}
sales = []

change = lambda total_bill, given_money : given_money - total_bill

def total_bill():
    try: 
        total = 0
        items = int(input("How many items do you have?: "))
        for item in range(items):
            item_price = float(input(f"Enter item{item +1} price:"))
            total += item_price
        return total
    except:
        return 0
        
def count_sales():
    for i in range(len(sales)):
        print(f"{i+1}. SAR{sales[i]}")
    print(f"Total: SAR {sum(sales)}")

cashier_name = input("What is your name?: ")
is_on = True
while is_on:
    print("\nType the number of the operation you want:")
    operation = input("1. Count customer items.\n2. Customer List.\n"
                      "3. check today sales\n4. Make a payment. \n5. Turn the machine off \n")

    if operation == "1":
        total = total_bill()
        if total == 0:
            print("Invalid value, Try Again")
        else:
            print(f"The total: SAR {total}\n")
            try:
                payment = float(input("How much do you want to pay?: "))
                sales.append(total)
                
                if payment < total:
                    name = input("What is your name?: ")
                    customer_info[name] = abs(change(total, payment))
                else:
                    print(f"Take SAR {change(total, payment)}\n")
            except:
                print("Invalid value, Try Again")
                
    elif operation == "2":
        print("customer List")
        for key, value in customer_info.items():
            print(f"Customer Name: {key}      Amount: SAR {value}")
            
    elif operation == "3":
        count_sales()
        
    elif operation == "4":
        name = input("What is your name?: ")
        payment = float(input("How much do you want to pay?: "))
        try:
            customer_info[name]-= payment
        except:
            print(f"There is no account for {name}")
        
    elif operation == "5":
        is_on = False
        message = lambda name: f"The Machine is off, Good bye {name}"
        print(message(cashier_name))
        
        
    else:
        print("Wrong choice, try again")
        print(sales)
```
